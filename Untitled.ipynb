```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


testing_gene_expr = pd.read_csv("10_RNAseq.csv", header=0, index_col=0)
testing_gene_expr = testing_gene_expr.transpose()
#remove col with treatment success
testing_gene_expr = testing_gene_expr.drop(testing_gene_expr.columns[-1], axis = 1)
```

```python
import seaborn as sns
```

```python
#analysis of dataset (survival)

#read dfs
training_clinic = pd.read_table("28_clinical_data.txt", header = 0)
training_clinic = pd.read_table("28_clinical_data.txt")
values = {'Patient ID': "Pt27B"}
training_clinic = training_clinic.fillna(value=values)
RNAseq = training_clinic['RNAseq'] == 1
RNAseq = RNAseq.tolist()
training_clinic = training_clinic[RNAseq]

training_response = pd.read_csv("aPD1_38.csv")
training_response = training_response.fillna(value=values)
training_response = training_response[RNAseq]
training_response["Response"] = training_response["Response"].map({"R":1, "NR":0})
training_response["Gender"] = training_response["Gender"].map({"F":1, "M":0})
training_response["Vital Status"] = training_response["Vital Status"].map({"Alive":1, "Dead":0})


training_gene_expr = pd.read_csv("28_RNAseq.csv", header = 0, index_col = 0)
training_gene_expr = training_gene_expr.transpose()

y_train = training_response['Response']
d = {'Response':[1, 1, 0, 0, 0]}
y_test = pd.DataFrame(data = d)
y_train = pd.DataFrame(data = y_train)
```

```python
#distribution of mean and median
training_gene_expr_log = np.log1p(training_gene_expr)
training_gene_expression_mean = training_gene_expr_log.mean()
training_gene_expression_median = training_gene_expr_log.median()

sns.distplot(training_gene_expression_mean)
plt.title("Distribution of mean expression")
plt.ylabel("Mean expression")
plt.show()
sns.distplot(training_gene_expression_median)
plt.title("Distribution of median expression")
plt.ylabel("Median expression")
plt.show()
```

```python
#getting rid of genes with low expression.1.5 is a random threshold
to_delete = []
for i in range(training_gene_expr_log.shape[1]):
    if sum(training_gene_expr_log.iloc[:, i]) < 1.5:
        to_delete.append(i)

training_gene_expr_log = training_gene_expr_log.drop(training_gene_expr_log.columns[to_delete], axis = 1)
```

```python
#now all low-expressed genes(the same as in training set) will be removed from testing
testing_gene_expr = testing_gene_expr.drop(testing_gene_expr.columns[to_delete], axis = 1)

#split testing dataset into baseline and after treatment
testing_baseline = testing_gene_expr[testing_gene_expr.index.str.contains("baseline")]
testing_baseline = testing_baseline.convert_objects(convert_numeric = True)

testing_after_treatment = testing_gene_expr[testing_gene_expr.index.str.contains("OnTx")]
testing_after_treatment = testing_after_treatment.convert_objects(convert_numeric = True)
```

```python
#In future, we are working with log data
X_train = training_gene_expr_log
testing_baseline_log = np.log1p(testing_baseline)
testing_baseline_log_mean = testing_baseline_log.mean()
testing_baseline_log_median = testing_baseline_log.median()
X_test = testing_baseline_log
```

```python
#check what is the difference in column names and fix some names
from sets import Set

#fix some colnames
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
names = X_train.columns.tolist()
for i in names:
    if i.split("-")[0] in months:
        names[names.index(i)] = i.split("-")[1] + "-" + i.split("-")[0]
X_train.columns = names

#see genes absent in one dataset
tr = Set(X_train.columns)
te = Set(X_test.columns)
diff = tr-te
diff2 = te-tr
```

```python
#drop duplicates
X_train = X_train.loc[:,~X_train.columns.duplicated()]
X_test = X_test.loc[:,~X_test.columns.duplicated()]

X_concat = pd.concat([X_train, X_test])
y_concat = pd.concat([y_train, y_test])

#let us delete all columns with NAs
X_concat = X_concat.dropna(axis = 1, how = "any")

y_concat = y_concat.reset_index(drop=True)
y_arr = y_concat.values
y_arr = y_arr.ravel()

y_tr = y_arr[:27]
y_te = y_arr[28:]

X_tr = X_concat.iloc[:27, :]
X_te = X_concat.iloc[28:, :]
```

```python
#labels for test set
labels = pd.DataFrame(data = X_te.index).values.ravel()

#PCA for merged data
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

pca = PCA(n_components=2)
X_cc = pca.fit(X_concat).transform(X_concat)

X_cc_tr = X_cc[:27, :]
X_cc_te = X_cc[28:, :]

plt.figure()
colors_tr = ['navy', 'turquoise']
colors_te = ["red", "green"]
lw = 2

for color, i in zip(colors_tr, [0, 1]):
    plt.scatter(X_cc_tr[y_tr == i, 0], X_cc_tr[y_tr == i, 1], color=color, alpha=.8, lw=lw)

for color, i in zip(colors_te, [0, 1]):
    plt.scatter(X_cc_te[y_te == i, 0], X_cc_te[y_te == i, 1], color=color, alpha=.8, lw=lw)
    

for label, x, y in zip(labels, X_cc_te[:, 0], X_cc_te[:, 1]):
    plt.annotate(
        label,
        xy=(x, y), xytext=(-20, 20),
        textcoords='offset points', ha='right', va='bottom',
        bbox=dict(boxstyle='round,pad=0.5', fc='yellow', alpha=0.5),
        arrowprops=dict(arrowstyle = '->', connectionstyle='arc3,rad=0'))


plt.show()
#they are mixed af, we can't analyze them because of partial identity of train and test. But we will.
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import average_precision_score
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
#try to predict test values just for fun
clf = RandomForestClassifier(n_estimators = 20)
classifier = clf.fit(X_tr, y_tr)
y_pred = clf.predict(X_te)

predictions = pd.DataFrame(data = classifier.predict_proba(X_te))[1]
print "predicted responce: \n", predictions
print "predicted classes: ", y_pred
```

```python
#which features were the most important?
import matplotlib.pyplot as plt

importances = clf.feature_importances_
features = X_tr.columns.values
imp_genes = sorted(zip(map(lambda x: round(x, 4), importances), features), 
             reverse=True)[:40]
#draw the barplot
height = []
genes = []
x = range(40)
for i in imp_genes:
    height.append(i[0])
    genes.append(i[1])


plt.plot(height, color = "green")
plt.title("Feature importances after filtering")
plt.xticks(x, genes, rotation='vertical', fontsize = 6)
#plt.rc(genes, fontsize = 1)#add xlabels genes
plt.show()
```

```python
#leave only the most important genes (40 genes shown above)
X_cc_filtered = X_concat.loc[:,genes]
X_cc_filt_tr = X_cc_filtered.iloc[:27, :]
X_cc_filt_te = X_cc_filtered.iloc[28:, :]
```

```python
X_cc_filt_tr = X_cc_filt_tr.dropna(axis=1)
X_cc_filt_te = X_cc_filt_te.dropna(axis=1)
```

```python
#logistic regression on filtered data
from sklearn import linear_model

logistic = linear_model.LogisticRegression(C = 0.1)
logistic.fit(X_cc_filt_tr, y_tr)
#training scores
print "log score: \n", logistic.score(X_cc_filt_tr, y_tr, sample_weight=None)
#predicted classes
predictions = pd.DataFrame(data = logistic.predict_proba(X_cc_filt_te))[1]
print "predicted responce: \n", predictions
```
